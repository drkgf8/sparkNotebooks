```
from pyspark.sql import SparkSession
import pandas as pd
```

```
spark = SparkSession.builder.appName('dates').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/0.csv',inferSchema=True,header=True)
```

```
from pyspark.sql.functions import dayofmonth as dom, dayofyear as doy, weekofyear as woy, date_format as dat_form
```

```
df.select(dom(df['birthDate'])).show()
```

```
from pyspark.sql.functions import month, year
```

```
df.select(year(df['birthDate'])).show()
```

```
from pyspark.sql.functions import year, hour, minute, format_number
```

```
df.select(hour(df['commitTimestamp'])).show()
```

```
df.select(minute(df['commitTimestamp'])).show()
```

```
new_df = df.withColumn("Year",year(df['commitTimestamp']))
```

```
result = new_df.groupBy('Year').mean()['avg(Year)','avg(age)']
```

```
result = result.withColumnRenamed('avg(Year)','Year')
```

```
result = result.select('Year',format_number('avg(age)',2).alias("Mean Age")).show()
```
