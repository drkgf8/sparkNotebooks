```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('missingdata').getOrCreate()
```

```
df = spark.read.csv("/FileStore/tables/0.csv",header=True,inferSchema=True)
```

```
df.show()
```

```
dfmin = df['id','commitTimestamp','email']
```

```
dfmin.show()
```

```
dfmin.na.drop().show()
```

```
dfmin.na.drop(thresh=2).show()
```

```
dfmin.na.fill('something@gmail.com').show()
```

```
dfmin.na.fill(0).show()
```

```
dfmin.na.fill('dkrounder@yahoo.com',subset=['email']).show()
```

```
df.printSchema()
```

```
dfmin = df['id','commitTimestamp','email','age']
```

```
dfmin.show()
```

```
from pyspark.sql.functions import mean
```

```
mean_val = dfmin.select(mean(df['age'])).collect()
```

```
mean_val
```

```
mean_age = mean_val[0][0]
```

```
dfmin.na.fill(mean_age,['age']).show()
```
