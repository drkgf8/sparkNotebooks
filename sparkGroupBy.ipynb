```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('gropubyapp').getOrCreate()
```

```
df = sqlContext.read.format("csv")\
  .option("header","true")\
  .option("inferSchema","true")\
  .load("/FileStore/tables/KansasCityChiefs{\w*}passRushRec.csv")
```

```
df.createOrReplaceTempView("dftemp")
```

```
sql_results = spark.sql("select Player, PassRate, PassYds from dftemp where PassYds > 100")
```

```
sql_results.groupBy("Player").mean().show()
```

```
sql_results.show()
```

```
from pyspark.sql.functions import countDistinct, avg, stddev
```

```
sql_results.select(countDistinct("Player")).show()
```

```
sql_results.select(avg("PassRate").alias("Average PassRate")).show()
```

```
from pyspark.sql.functions import format_number
```

```
aver =  sql_results.select(avg("PassRate").alias("Average PassRate"))
```

```
aver.select(format_number("Average PassRate",2).alias("Average Pass Rate")).show()
```

```
sql_results.orderBy("PassRate").show()
```

```
sql_results.orderBy(sql_results["PassRate"].desc()).show()
```
